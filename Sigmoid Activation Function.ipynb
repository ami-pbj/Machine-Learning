```python
import math
```

**Sigmoid Function**

*To plot a Sigmoid Activation Function, I need to write a sub function to return y value for each x input.*

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

**To Plotting Sigmoid Activation Function**

*Importing Numpy and Matplotlib* 

```python
import numpy as np
import matplotlib.pyplot as plt
```

*Create an array named x between -10 to 10 with an interval of 0.2,*
*then pass the array x into sigmoid() which returns an array with corresponding values.*

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
```

**Calling Plot**

```python
plt.plot(x,sig)
plt.show()
```
