## **Data Ready for Training Linear Regression**

### **X and y List - Partition**

```python
X = HousePrice[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated']]

y = HousePrice['price']
```

### **Split Data Into Train, Test**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

### **Creating and Training Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression 

lm = LinearRegression() 

lm.fit(X_train,y_train) 
```

### **LR** **Model** **Evalution**

```python
print(lm.intercept_)
```

```python
coefficients = pd.DataFrame(lm.coef_, X.columns)
coefficients.columns = ['Coefficients']
# coefficients = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient']) 

# To showing output
coefficients
```

# **Prediction from Linear Regression Model**

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```
